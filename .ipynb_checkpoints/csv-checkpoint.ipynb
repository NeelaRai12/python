```python
info = ""
n = int(input("Enter n = "))
for i in range(n):
    name = input("Enter name = ")
    lname = input("Enter lname = ")
    physics = int(input("Enter physics = "))
    chemistry = int(input("Enter chemistry = "))
    math = int(input("Enter math = "))
    english = int(input("Enter English = "))
    total = physics+chemistry+math+english
    per = total/4
    data = f"{name},{lname},{physics},{chemistry},{math},{english},{total} ,{per}\n"
    info = info +data
print(info)
b = open('data.csv','a')
b.write('Name,Lname,Physics,Chemistry,Math,English,Total,Per\n')
b.write(info)
b.close()
```

# Concate

```python
info = {"Name":["Asha","Jeetan","Neela"],"Dept":["Civil","Cullinary","Comuter"],"Salary":[40000,30000,50000]}
```

```python
import pandas as pd
df1 = pd.DataFrame(info)
df1
```

```python
info = {"Name":["Asha","Jeetan","Neela"],"Post":["Sub-engineer","Chef","Manager"]}
```

```python
import pandas as pd
df2 = pd.DataFrame(info)
df2
```

```python
df3 = pd.merge(df1,df2,on="Name")
df3
```

```python
details = {"City":["Ktm","Pkr","Bkt","Lalitpur"],"Temperature":[21,20,25,30]}
```

```python
import pandas as pd
df1 = pd.DataFrame(details)
df1
```

```python
details = {"City":["Ktm","Pkr","Bkt","Brt"],"Humidity":[65,58,67,74]}
```

```python
import pandas as pd
df2 = pd.DataFrame(details)
df2
```

```python
df3 = pd.merge(df1,df2,on="City")
df3
```

```python
df3 = pd.merge(df1,df2,how="outer")
df3
```

```python
df3 = pd.merge(df1,df2,how="inner")
df3
```

```python
df3 = pd.merge(df1,df2,how="left")
df3
```

```python
df3 = pd.merge(df1,df2,how="right")
df3
```

```python
df3 = pd.merge(df1,df2,how="outer",indicator='True')
df3
```
